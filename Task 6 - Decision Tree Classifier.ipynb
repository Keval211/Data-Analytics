```python
#import the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
#Load the Dataset of Iris
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
#Remove the unnecessary columns
df = df.drop(["Id"],axis=1)
```

```python
#Distrbute the dataset into input dataset & target dataset
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
#Split the dataset into train & test dataset wherein test size=0.2 means train dataset  is 80% of total dataset
x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#defining the decision tree algorithm model
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
#Predict the target for the input test dataset
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
x1 = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
y1 = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
```

```python
#Display the Tree diagram
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)

tree.plot_tree(clf, feature_names = x1, class_names=y1, filled = True)
```

```python
#Check the decision tree classifier model for the below check input values
check = pd.DataFrame([[1.2,2.1,2.2,3.1]])
check
```

```python
#Predict the target values for the given check input values
answer = clf.predict(check)
answer
```
